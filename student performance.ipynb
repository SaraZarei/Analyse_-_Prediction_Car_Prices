<h2>Statistical Tests</h2>

<h3>T-test (two sample t-test, one tailed)</h3>
<p><b>Research question</b>: Is the difference between male average math score and female average math score is significant?</p>
<p><b>H0</b>: male average math score = female average math score</p>
<p><b>Ha</b>: male average math score != female average math score</p>

<h3>Checking T-test assumptions </h3>

<p><b>Normality of data</b></p>

```python
female_math=df.loc[df['gender']=='female','math score']
male_math=df.loc[df['gender']=='male','math score']
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=1,ncols=2)
sns.distplot(female_math,ax=ax0)
sns.distplot(male_math,ax=ax1)
ax0.set(title='Female')
ax1.set(title='Male')
```

<p>Two groups are approximately normall</p>

<p><b>Hemogenity of variance</b></p>

```python
print('female variance:',female_math.std())
print('male variance:',male_math.std())
```

<p>Two groups have approximately same variance</p>

<p><b>Test statistic and P-value</b></p>

```python
from scipy.stats import ttest_ind 
statistic,p_value=ttest_ind(female_math,male_math)
print('test_statistic:',statistic,'\t','p_value:',p_value)
```

<p>This means the probability to see this test statistic (5.38 difference between average male math score and average female math score) by chance is 9.120185549328822e-08.So,we reject the null hypothesis </p>

<p><b>Inference</b></p>
<p>The difference in math score between male(Mean=68, SD=14.35) and female(Mean=63.5, SD=15.49) is significant(test_statistic: -5.38, p_value &#60; 0.05</p>
 

<h3>T-test (two sample t-test, one tailed)</h3>
<p><b>Research question</b>: Is the difference between male average reading score and female average reading score is significant?</p>
<p><b>H0</b>: male average reading score = female average reading score</p>
<p><b>Ha</b>: male average reading score != female average reading score</p>

```python
female_reading=df.loc[df['gender']=='female','reading score']
male_reading=df.loc[df['gender']=='male','reading score']
fig,(ax0,ax1)=plt.subplots(nrows=1,ncols=2)
sns.distplot(female_reading,ax=ax0)
sns.distplot(male_reading,ax=ax1)
ax0.set(title='Female')
ax1.set(title='Male')
```

<p>Two groups are approximately normall</p>

```python
print('female variance:',female_reading.std(),'male variance:',male_reading.std(),'\n')

statistic,p_value=ttest_ind(female_reading,male_reading)
print('test_statistic:',statistic,'\t','p_value:',p_value)
```

<p>Two groups have approximately same variance</p>
<p>This means the probability to see this test statistic (7.95 difference between average male reading score and average female reading score) by chance is  4.680538743933289e-15.So,we reject the null hypothesis </p>
<p><b>Inference</b></p>
<p>The difference in reading score between male(Mean=65, SD=13.93) and female(Mean=73, SD=14.37) is significant(test_statistic: 7.95, p_value &#60; 0.05</p>

<h3> one-way ANOVA + Tukeys Test  </h3>
<p><b>Research question</b>:What is difference in average math score  among ethnicity groups ?</p>
<p><b>H0</b>: There is no difference between these groups</p>
<p><b>Ha</b>: There is difference between these groups</p>

```python
math_A=df.loc[df['race/ethnicity']=='group A','math score']
math_B=df.loc[df['race/ethnicity']=='group B','math score']
math_C=df.loc[df['race/ethnicity']=='group C','math score']
math_D=df.loc[df['race/ethnicity']=='group D','math score']
math_E=df.loc[df['race/ethnicity']=='group E','math score']
```

```python
from scipy.stats import f_oneway
statistic,p_value=f_oneway(math_A,math_B,math_C,math_D,math_E)
print('test_statistic:',statistic,'\t','p_value:',p_value)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey=pairwise_tukeyhsd(endog=df['math score'],groups=df['race/ethnicity'],alpha=0.05)
print(tukey)
```

<p><b>Inference</b></p>
<p>We found a statistically significant difference in average math score according to ethnicity (f-test=14.59, p&#60;0.05)</p>
<p>A Tukey post-hoc test revealed significant pairwise differences between group A and E with an average difference of 12.19,between group A and D with an average difference of 5.73,between group B and D with an average difference of 3.91,between group B and E with an average difference of 10.36,between group C and E with an average difference of 9.35,between group D and E with an average difference of 6.45. Between other groups is not statistically significant difference</p>

Between other groups is not statistically significant difference</p>

<h3>check for hemoscedasticity condition</h3>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import train_test_split
x=df[['lunch']]
y=df['overal score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
lr=LinearRegression()
```

```python
r_squared=cross_val_score(lr,x_train,y_train,cv=5)
print('R-squared for every fold:',r_squared)
print(' Average R-squared:',np.mean(r_squared))
```

```python
lr.fit(x_train,y_train)
y_predict=lr.predict(x_test)
residual=y_test - y_predict
fig,ax=plt.subplots(figsize=(6,6))
sns.regplot(x=y_predict,y=residual,ax=ax)
ax.set_title('plot of residuals vs predicted values for lunch feature',y=1.05)
```

```python
print('MSE:',mean_squared_error(y_predict,y_test))
```

<h4><p>As seen R-squared is very low and MSE is very high, we know a good linear model has to have high R-squared and low </p> <p>errors(MSE).In the other hand,graph of residuals versus predicted values isn't spread symetricaly around the regline, so it </p> <p>violates hemoscedasticity condition for regression(the variance in the residuals is assumed to be constant)</p> 
</h4>
